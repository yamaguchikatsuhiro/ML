```python
df_2 = df.drop_duplicates(keep="first").reset_index(drop=True)
df_2.shape
```

```python
df_2 = df_2.drop(["selector"], axis=1)

df_2.info()
```

```python
X = df_2.drop(["drinks"], axis=1)
y = df_2["drinks"]
```

```python
df_2["gammagt"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=.2,
                                                    shuffle=True,
                                                    random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
```

```python
sc.fit(X_train, y_train)
```

```python
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)
```

```python
import lightgbm as lgb
from sklearn.model_selection import KFold
from sklearn.metrics import mean_absolute_error

valid_scores = []
models = []
oof = np.zeros(len(X_train))
```

```python
params = {
    "objective": "regression_l1",
    "seed": 0,
    "verbose": -1,
    "learning_rate": .1,
}
```

```python
kf = KFold(n_splits=5, shuffle=True, random_state=0)

for fold, (tr_idx, va_idx) in enumerate(kf.split(X=X_train_std, y=y_train)):
    X_tr_std = X_train_std[tr_idx, :]
    y_tr = y_train.iloc[tr_idx].values
    X_va_std = X_train_std[va_idx, :]
    y_va = y_train.iloc[va_idx].values

    lgb_train = lgb.Dataset(X_tr_std, y_tr)
    lgb_eval = lgb.Dataset(X_va_std, y_va, reference=lgb_train)

    model = lgb.train(params,
                      lgb_train,
                      valid_sets=[lgb_train, lgb_eval],
                      valid_names=["train", "valid"],
                      num_boost_round=10000,
                      callbacks=[lgb.early_stopping(100),
                                 lgb.log_evaluation(500)])

    y_va_pred = model.predict(X_va_std, num_iteration=model.best_iteration)
    score = mean_absolute_error(y_va, y_va_pred)

    print(f"fold:{fold+1}, MAE valid: {score:.3f}")
    print("-------------------------------------------------------------------")

    valid_scores.append(score)
    models.append(model)
    oof[va_idx] = y_va_pred


```

```python
cv_score = np.mean(valid_scores)
print(f"CV score: {cv_score:.3f}")
```
