```python
df = df.drop(["veil-type"], axis=1).reset_index(drop=True)
```

```python
df = df.astype("category")
df_ohe = df.copy()
df_le = df.copy()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

for col in df_le.columns:
    df_le[col] = le.fit_transform(df_le[col])
```

```python
import lightgbm as lgb
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.metrics import log_loss, accuracy_score
```

```python
X = df_le.drop(["poisonous"], axis=1)
y = df_le["poisonous"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=.2,
                                                    random_state=0,
                                                    stratify=y)
```

```python
params = {
    "objective": "binary",
    "num_leaves": 5,
    "seed": 0,
    "verbose": -1
}

accuracy_scores = []
logloss_scores = []
models = []
oof = np.zeros((len(X_train, )))
```

```python
kf = StratifiedKFold(n_splits=5, shuffle=True, random_state=0)

for fold, (tr_idx, va_idx) in enumerate(kf.split(X=X_train, y=y_train)):
    X_tr = X_train.iloc[tr_idx, :]
    y_tr = y_train.iloc[tr_idx]
    X_va = X_train.iloc[va_idx, :]
    y_va = y_train.iloc[va_idx]

    lgb_train = lgb.Dataset(X_tr, y_tr)
    lgb_eval = lgb.Dataset(X_va, y_va, reference=lgb_train)

    model = lgb.train(params,
                      train_set=lgb_train,
                      num_boost_round=500,
                      valid_sets=[lgb_train, lgb_eval],
                      valid_names=["train", "valid"],
                      callbacks=[lgb.early_stopping(10),
                                 lgb.log_evaluation(100)])
    y_va_pred = model.predict(X_va, num_iteration=model.best_iteration)

    accuracy = accuracy_score(y_va, np.round(y_va_pred))
    logloss = log_loss(y_va, y_va_pred)

    print(f"fold: {fold + 1}, accuracy_score: {accuracy:.4f}, log_loss: {logloss:.4f}")

    print("----------------------------------------------------------------------------")

    accuracy_scores.append(accuracy)
    logloss_scores.append(logloss)
    models.append(model)
    oof[va_idx] = y_va_pred

    cv_accuracy_score = np.mean(accuracy_scores)
    cv_logloss_score = np.mean(logloss_scores)

    print(f"CV accuracy score: {cv_accuracy_score:.4f}, CV logloss score: {cv_logloss_score:.4f}")

```
