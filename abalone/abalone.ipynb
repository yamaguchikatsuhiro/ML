```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data_url = "https://archive.ics.uci.edu/static/public/1/data.csv"
```

```python
df = pd.read_csv(data_url)
df.head()
```

```python
df.info()
```

```python
df.columns
```

```python
df[df.duplicated(keep=False)]
```

```python
df["Sex"] = df["Sex"].astype("category")
df["Sex"].unique()
```

```python
df.describe()
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(sparse_output=True)

ohe = ohe.fit_transform(df[["Sex"]])
df_dummy = pd.DataFrame(ohe.toarray(),
                        columns=["Sex_" + str(col) for col in df["Sex"].unique()])
```

```python
df_dummy.head()
```

```python
df = pd.concat([df, df_dummy], axis=1).drop("Sex", axis=1)
df.head()
```

```python
df.info()
```

```python
from sklearn.model_selection import train_test_split

X = df.drop(["Rings"], axis=1)
y = df["Rings"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, shuffle=True, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
sc.fit(X_train)

X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)
```

```python
import lightgbm as lgb
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error

valid_scores = []
models = []
oof = np.zeros(len(X_train))

params = {
    "objective": "regression_l2",
    "metrics": "rmse",
    "boosting": "gbdt",
    "learning_rate": .01
}
```

```python
kf = KFold(n_splits=5,
           shuffle=True,
           random_state=0)

for fold, (tr_idx, va_idx) in enumerate(kf.split(X=X_train_std)):
    X_tr = X_train_std[tr_idx, :]
    y_tr = y_train.iloc[tr_idx]
    X_va = X_train_std[va_idx, :]
    y_va = y_train.iloc[va_idx]

    lgb_train = lgb.Dataset(X_tr, y_tr)
    lgb_eval = lgb.Dataset(X_va, y_va, reference=lgb_train)

    model = lgb.train(params,
                      lgb_train,
                      num_boost_round=10000,
                      valid_sets=[lgb_train, lgb_eval],
                      valid_names=["train", "valid"],
                      callbacks=[lgb.early_stopping(100),
                                 lgb.log_evaluation(500)])

    y_va_pred = model.predict(X_va, num_iteration=model.best_iteration)

    score = np.sqrt(mean_squared_error(y_va, y_va_pred))
    print(f"fold:{fold+1}, RMES valid:{score:.3f}")
    print("------------------------------------------------------------")

    valid_scores.append(score)
    models.append(model)
    oof[va_idx] = y_va_pred

```

```python
cv_score = np.mean(valid_scores)
print(f"CV score:{cv_score:.3f}")
```
